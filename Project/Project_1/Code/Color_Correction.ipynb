```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
link_img = "..\\Image_Input\\conan.jpg"
```

```python
def convert_RGB_to_gray(image) :
    gray_image = []
    for row in image:
        gray_row = []
        for pixel in row:
            r, g, b = pixel
            gray = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
            gray_row.append(gray)
        gray_image.append(gray_row)
    return gray_image
```

```python
def reversed_black_white_color(image) :
    return [[255 - pixel for pixel in row] for row in image]
```

```python
if __name__ == '__main__' :
    img = cv2.imread(link_img)
    img_gray = convert_RGB_to_gray(img)
    img_reversed = reversed_black_white_color(img_gray)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Gray Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(img_reversed, cmap='gray')
    plt.title('Reversed Image')
    plt.axis('off')

    plt.show()
```
