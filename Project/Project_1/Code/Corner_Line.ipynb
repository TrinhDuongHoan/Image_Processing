```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
link_img = "..\\Image_Input\\conan.jpg"
```

```python
img = cv2.imread(link_img)
img = cv2.resize(img, (500, 500))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def create_corner_line(image) : 
    for i in range (100) :
        for j in range (100) :
            if i + j  == 100 :
                for jj in range(50) :
                    image[i ][j + jj] = 0
                    image[i + jj][j] = 0 
    return image
                
```

```python
new_img = create_corner_line(img)   
plt.title("Corner Line Image")
plt.imshow(new_img)
plt.axis("off")
```
