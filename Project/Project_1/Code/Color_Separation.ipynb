```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
link_img = "..\\Image_Input\\black_white_image.jpg"
```

```python
def threshold_image(img_gray, threshold):
    img_thresholded = np.copy(img_gray)
    
    img_thresholded[img_thresholded < threshold] = 0
    
    return img_thresholded

def apply_thresholds(img_gray, thresholds):

    plt.figure(figsize=(15, 5))
    
    for i, threshold in enumerate(thresholds):
        img_thresholded = threshold_image(img_gray, threshold)
        plt.subplot(1, 3, i+1)
        plt.imshow(img_thresholded, cmap='gray')
        plt.title(f"Threshold: {threshold}")
        plt.axis('off')
    
    plt.show()


img_gray = cv2.imread(link_img, cv2.IMREAD_GRAYSCALE)
thresholds = [32, 64 , 127]
apply_thresholds(img_gray, thresholds)
```
