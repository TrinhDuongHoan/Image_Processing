```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
link_img1 = "..\\Image_Input\\test3.png"
link_img2 = "..\\Image_Input\\test4.png"
```

```python
def find_secret_by_subtract(image1, image2) :
    return image2 - image1
```

```python
img1 = cv2.imread(link_img1, 0)
img2 = cv2.imread(link_img2, 0)

img_sub = find_secret_by_subtract(img1, img2)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Image 1')
plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title('Image 2')
plt.subplot(1, 3, 3)
plt.imshow(img_sub, cmap='gray')
plt.title('Image Subtract')
plt.show()


```
