```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Task 1

```python
link_img = "..\\LabWeek\\anhconan.jpg"
```

```python
img = cv2.imread(link_img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
width, height = img.shape[:2]
plt.imshow(img)
plt.axis("off")
```

#### Task 2

```python
def flip_horizontal(img) :
    f_img = img.copy()
    w, h = f_img.shape[:2]

    for i in range(h) :
        for j in range(w) :
            f_img[h - i - 1][j] = f_img[i][j]
    return f_img


flip_img = flip_horizontal(img)
plt.figure()
plt.imshow(flip_img)
plt.axis("off")
```

#### Task 3

```python
def create_padding(pad_size, img_size) :
    res = np.full((img_size + 2 * pad_size,img_size + 2 * pad_size,3), 0, dtype=np.uint8)
    res[pad_size:pad_size + img_size, pad_size:pad_size +img_size] = 255
    return res

pad = create_padding(10, 500)
plt.axis("off")
plt.imshow(pad)
```

#### Task 4

```python
def create_rectangle(ww) :
    list_toado = [[0,0], [0, height-ww], [width-ww  , 0], [width-ww- 1, height-ww ]]
    r_img = np.full((width, height, 3), 255, dtype = np.uint8)
    for i in list_toado :
        x, y = i
        r_img[x:x+ww, y:y+ww] = 0
    return r_img

def make_image(thickness, d) :
    res_img = np.full((300, 300, 3), 255, dtype = np.uint8)
    
    for i in range(0, 300, d) :
        res_img[i:i+thickness, :] = [0, 0, 0]
        res_img[:, i:i+thickness] = [0, 0, 0]
    return res_img

def create_square(ww) :
    s_img = np.zeros((width, height), dtype = np.uint8)
    start_center = [width//2, height//2]
    x, y = start_center
    x, y = x - ww//2, y - ww//2
    s_img[x:x+ww, y:y+ww] = 255
    return s_img
```

```python
sq_img = create_square(100)
rec_img = create_rectangle(50)
caro_img = make_image(10, 30)

plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(sq_img, cmap = "gray")
plt.title("Square Image")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(rec_img, cmap = "gray")
plt.title("Rectangle Image")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(caro_img)
plt.title("Caro Image")
plt.axis("off")
```

```python
new_img = cv2.imread(link_img)
new_img = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
print(new_img.shape)
plt.imshow(new_img)
```

```python
up_img = cv2.resize(new_img, (512, 512))
plt.imshow(up_img)
```

```python
def down_size(img) :    
    temp = np.zeros((img.shape[0]//4, img.shape[1]//4, 3), dtype = np.uint8)
    for i in range(0, img.shape[0], 4) :
        for j in range(0, img.shape[1], 4) :
            temp[i//4][j//4] = img[i][j]
    return temp
```

```python
down_img = down_size(new_img)
plt.axis("off")
plt.imshow(down_img)
plt.savefig("output5.jpg")
```

```python
def create_4color_rectangle(ww) :
    list_toado = [[0,0], [0, height-ww], [width-ww  , 0], [width-ww- 1, height-ww ]]
    r_img = np.full((width, height, 3), 255, dtype = np.uint8)
    list_color = [[255, 0, 0], [0, 255, 0], [0, 0, 255], [0, 0, 0]]
    for i in list_toado :
        x, y = i
        r_img[x:x+ww, y:y+ww] = list_color[list_toado.index(i)]
    return r_img
```

```python
c_img = create_4color_rectangle(50)
plt.imshow(c_img)
```
